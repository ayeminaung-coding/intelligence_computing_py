- **gpa** (float)

### 2. Grade Evaluation
Determine the student's **grade** based on the score using the criteria below:

| Score Range | Grade |
|------------|-------|
| ‚â• 80       | A     |
| ‚â• 70       | B     |
| ‚â• 60       | C     |
| ‚â• 50       | D     |

```python
student_id = input("Enter Student ID: ")
# print(f"Your student ID is: {student_id}")
student_name = input("Enter Student Name: ")
print(f"Enter name: {student_name}")
student_age = input("Enter age: ")
print(f"Enter age : {student_age}")
student_gpa = input("Enter GPA: ")
print(f"Enter GPA : {student_gpa}")
student_score = input("Enter score: ")
print(f"Enter score (0-100): {student_score}")
print("------------------------------")
print("The result Information")
print("------------------------------")

def calculate_grade(score):
    try:
        score = float(score)
        if score >= 90:
            return 'A'
        elif score >= 80:
            return 'B'
        elif score >= 70:
            return 'C'
        elif score >= 60:
            return 'D'
        else:
            return 'F'
    except ValueError:
        return 'Invalid score'

print(f"Student ID: {student_id}")
print(f"Name: {student_name}")
print(f"Age: {student_age}")
print(f"GPA: {student_gpa}")
print(f"Score: {calculate_grade(student_score)}")
```

## üìù Question#2: List Cleaning and Summary

Write a Python program that performs the following tasks:
### 1. Input
- Input **8 integers** from the keyboard (one by one).
- Store all input values in a **list**.
---
### 2. List Processing
- Create a **new list** that contains only numbers from the original list that:
  - are **positive number** and **divisible by 2**
---
### 3. Output
- Display the new filtered list.
- Display the **sum** of the numbers in the filtered list.
- Display the **maximum value** in the filtered list.



```python
user_data = []

for i in range(8):
    item = input(f"Enter item {i+1}: ")
    user_data.append(item)
print("------------------------------")
print("The result Information")
print("------------------------------")

# create new list that contains positive number and divisible by 2 from original list
new_list = []

for num in user_data:
    if num.isdigit():
        converted_num = int(num)
        
        if converted_num > 0 and converted_num % 2 == 0:
            new_list.append(converted_num)
print("List number: ", new_list)

def calculate_sum(numbers):
    total = 0
    for number in numbers:
        total += number
    return total
print("Summation :", calculate_sum(new_list))

print("Maximun Value: ", max(new_list))
```

## üìù Question #3: Simple Student Database

Create a Python program that manages a simple student score database using a dictionary.

---
### 1. Input
- Allow the user to input data for **3 students**.
- For each student, receive:
  - **Student Name** (string)
  - **Student Score** (integer)
Store the data in a dictionary where:
- **Key** = Student ID  
- **Value** = Student Score  
---
### 2. Processing
Your program must:
- Display all student IDs and their scores
- Determine the **highest score**
- Identify the **student ID** with the highest score
---
### 3. Output
Display:
- The student Name that has the highest score


```python
student_records = {}

for i in range(3):
    name = input(f"Enter Student Name {i+1}:" )
    score = int(input(f"Enter {name}'s Score: "))
    print(f"Enter Student Name {i+1}: ", name)
    print(f"Enter score for {name}: ", score)
    
    student_records[name] = score
print("------------------------------")
print("\n--- The result information ---")
print("------------------------------")

def get_highest_scorer(records):
    highest_score = -1
    top_student = ""
    for student, score in records.items():
        if score > highest_score:
            highest_score = score
            top_student = student
    return top_student, highest_score

top_student, highest_score = get_highest_scorer(student_records)
print(f"Student with highest score: {top_student}\n Highest Score: {highest_score}")


```

An ATM allows users to withdraw money under the following rules:

A withdrawal is allowed only if:

- The **remaining balance** (if approved)

* Example#1

* Example#2

```python
account_balance = int(input("Enter account balance: "))
print(f"Enter Account balance: {account_balance}")
withdraw_amount = int(input("Enter withdraw amount: "))
print(f"Enter Withdraw amount: {withdraw_amount}")
fee_charge = 20

if withdraw_amount <= 0:
    print("Withdrawal denied: The amount must be greater than 0.")

elif withdraw_amount % 100 != 0:
    print("Withdrawal denied: The amount must be in ONLY 100 banknotes.")

elif withdraw_amount > account_balance:
    print(f"Withdrawal denied: please withdraw under the balance of {account_balance} baht")

elif withdraw_amount + fee_charge > account_balance:
    print("Withdrawal denied: Balance not enough for service fee")

else:
    new_balance = account_balance - withdraw_amount - fee_charge
    print("Withdrawal approved.")
    print(f"Remaining balance: {new_balance}")
```

Create **2 functions** as follows:
